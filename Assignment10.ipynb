```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# creat an empty undirected plot
G = nx.Graph()
# Add nodes

G.add_node("Sparsh")
G.add_nodes_from(["Haoyu", "Xiaoran", "Shuyi", "Jingyi"])

# Add edges
G.add_edge("Sparsh", "Haoyu")
G.add_edges_from([("Haoyu", "Xiaoran"), ("Xiaoran", "Shuyi"), ("Shuyi", "Jingyi"),
                 ("Jingyi", "Haoyu"), ("Sparsh", "Jingyi"), ("Sparsh", "Shuyi"),
                  ("Sparsh", "Xiaoran"),("Jingyi", "Xiaoran"), ("Shuyi", "Haoyu")])

# Plot network diagram
plt.figure(figsize=(7.5,7.5))
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
list(G.nodes)
```

```python
member_role = ["leader", "member", "member", "member", "member"]
for idx, node_id in enumerate(G.nodes):
    G.nodes[node_id]["role"] = member_role[idx]
```

```python
G.nodes["Shuyi"]
```

```python
node_color = [
    '#1f78b4' if G.nodes[v]["role"] == "member"
    else '#33a02c' for v in G]
```

```python
plt.figure(figsize=(7.5,7.5))
nx.draw_networkx(G, label = True, node_color=node_color)
plt.show()
```
